```python
from TextReader import TextReader
import nltk
import math
import random
import statistics as st
```

```python
all_texts = []
for i in range(1, 11):
    all_texts.append(TextReader().getText(f"{i}.txt"))
```

```python
all_texts
```

```python
all_words = []
for i in range(10):
    all_words.extend(all_texts[i])
    
print(len(all_words))
```

```python
unique_words = set(all_words)
print(len(unique_words))
```

```python
TF = []
for text in all_texts:
    words = dict()
    for w in unique_words:
        words[w] = text.count(w) / len(text)
    TF.append(words)

print(TF)
```

```python
IDF = dict()
number_of_documents = 4
for word in unique_words:
    counter = 1
    for text in all_texts:
        if word in text:
            counter += 1
    IDF[word] = math.log(10/counter)

print(IDF)
```

```python
TF_IDF = []
for tf in TF:
    result = []
    for word in unique_words:
        result.append(tf[word] * IDF[word])
    TF_IDF.append(result)
    
print(TF_IDF)
```

```python
def get_distance(list1,list2):
    eucl_distance = math.sqrt(sum((i - j)**2 for i,j in zip(list1,list2)))
    return eucl_distance
```

```python
def get_centroid(collection):
    centr_coord = []
    dim = len(collection[0])
        
    for i in range(dim):
        each_dim_coord = [point[i] for point in collection]
        each_dim_mean = st.mean(each_dim_coord)
        centr_coord.append(each_dim_mean)
    return centr_coord
```
